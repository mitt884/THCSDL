```python
import random
from faker import Faker

fake = Faker()

def generate_book_data(num_records):
    categories = range(1, 13)  # category_id from 1 to 12
    publishers = range(1, 23)  # publisher_id from 1 to 22
    
    books = []
    for _ in range(num_records):
        title = fake.sentence(nb_words=random.randint(5, 10), variable_nb_words=True).replace('.', '')
        published_year = random.randint(2007, 2022)  # Updated year range
        cost = random.randint(50000, 300000)
        author = fake.name()
        pages = random.randint(1, 999)
        current_number_of_copies = random.randint(1, 499)
        category_id = random.choice(categories)
        publisher_id = random.choice(publishers)
        
        books.append((title, published_year, cost, author, pages, current_number_of_copies, category_id, publisher_id))
    return books

def main():
    num_records = 9000
    book_data = generate_book_data(num_records)
    
    # Print the tuples
    for i, data in enumerate(book_data):
        title, published_year, cost, author, pages, current_number_of_copies, category_id, publisher_id = data
        print(f"('{title}', {published_year}, {cost}, '{author}', {pages}, {current_number_of_copies}, {category_id}, {publisher_id}),")

if __name__ == "__main__":
    main()
```
