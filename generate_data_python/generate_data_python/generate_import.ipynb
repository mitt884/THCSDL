```python
import random
import datetime

def generate_date(start, end):
    return start + datetime.timedelta(days=random.randint(0, (end - start).days))

def main():
    start_import_date = datetime.date(2021, 1, 1)
    end_import_date = datetime.date(2024, 3, 31)

    confirmer_ids = [f"NV{str(i).zfill(8)}" for i in range(1, 2001)]
    isbns = [f"{str(i).zfill(10)}" for i in range(1, 9001)]

    tuples = []
    
    for _ in range(10000):
        import_date = generate_date(start_import_date, end_import_date)
        number_of_copies = random.randint(50, 100)

        tuple_str = f"({number_of_copies}, '{import_date.strftime('%Y-%m-%d')}', '{random.choice(confirmer_ids)}', '{random.choice(isbns)}'),"
        tuples.append(tuple_str)

    for t in tuples:
        print(t)

if __name__ == "__main__":
    main()
```
