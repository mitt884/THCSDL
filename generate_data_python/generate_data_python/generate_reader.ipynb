```python
import random
from datetime import datetime, timedelta

# List of Vietnamese full names without accents
def generate_vietnamese_name():
    first_names = ["Nguyen", "Tran", "Le", "Pham", "Hoang", "Huynh", "Phan", "Vu", "Vo", "Dang", "Tran", "Dinh", "Trinh", "Ngo", "Ly"]
    middle_names = ["Van", "Thai", "Duc", "Minh", "Quoc", "Cong", "Huy", "Hong", "Bich", "Thanh", "Dieu", "Thanh", "An", "Hai", "Ngoc", "Thu"]
    last_names = ["An", "Binh", "Chau", "Dai", "Dung", "Hanh", "Hien", "Khanh", "Lam", "Mai", "Linh", "Trang", "Phuong", "Dat", "Tien", "Thao"]
    
    return f"{random.choice(first_names)} {random.choice(middle_names)} {random.choice(last_names)}"

# List of possible genders
genders = ['Nam', 'Nu']

# Function to generate a random date of birth between 1980 and 2006
def generate_dob():
    start_date = datetime(1980, 1, 1)
    end_date = datetime(2006, 12, 31)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    return random_date.strftime('%Y-%m-%d')

# List of possible email domains
email_domains = ['@gmail.com', '@hotmail.com', '@outlook.com']

# Function to generate a random email address
def generate_email(full_name):
    full_name = full_name.replace(' ', '').lower()
    domain = random.choice(email_domains)
    return f'{full_name}{domain}'

# Function to generate a random phone number (10 or 11 digits)
def generate_phone_number():
    return ''.join(random.choice('0123456789') for _ in range(random.randint(10, 11)))

# Generate a list of account_ids

# Generating tuples
for i in range(1, 8001):
    full_name = generate_vietnamese_name()
    gender = random.choice(genders)
    dob = generate_dob()
    email = generate_email(full_name)
    phone_number = generate_phone_number()
    
    
    # Print the tuple in the SQL INSERT format, each tuple on a new line
    print(f"('{full_name}', '{gender}', '{dob}', '{email}', '{phone_number}', {i}),")
```
