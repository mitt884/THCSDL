```python
import random
import datetime

def generate_date(start, end):
    return start + datetime.timedelta(days=random.randint(0, (end - start).days))

def main():
    start_borrow_date = datetime.date(2023, 11, 1)
    end_borrow_date = datetime.date(2024, 6, 30)
    end_actual_return_date = datetime.date(2024, 3, 31)

    reader_ids = [f"DG{str(i).zfill(8)}" for i in range(1, 8001)]
    isbns = [f"{str(i).zfill(10)}" for i in range(1, 9001)]

    tuples = []
    
    for _ in range(10000):
        borrow_date = generate_date(start_borrow_date, end_borrow_date)
        initial_due_date = borrow_date + datetime.timedelta(days=90)

        if borrow_date < datetime.date(2024, 3, 1):
            actual_return_date = generate_date(datetime.date(2024, 1, 1), end_actual_return_date)
            if actual_return_date < borrow_date:
                actual_return_date = borrow_date + datetime.timedelta(days=random.randint(1, 60))
            actual_return_date_str = f"'{actual_return_date.strftime('%Y-%m-%d')}'"
        else:
            actual_return_date_str = 'NULL' if random.random() < 0.9 else f"'{(borrow_date + datetime.timedelta(days=random.randint(91, 120))).strftime('%Y-%m-%d')}'"

        number_of_copies = random.randint(1, 5)
        tuple_str = f"('{borrow_date.strftime('%Y-%m-%d')}', '{initial_due_date.strftime('%Y-%m-%d')}', {actual_return_date_str}, {number_of_copies}, '{random.choice(reader_ids)}', '{random.choice(isbns)}'),"
        tuples.append(tuple_str)

    for t in tuples:
        print(t)

if __name__ == "__main__":
    main()
```
