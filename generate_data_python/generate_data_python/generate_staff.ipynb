```python
import random
import string
import datetime

def generate_vietnamese_name():
    first_names = ["Nguyen", "Tran", "Le", "Pham", "Hoang", "Huynh", "Phan", "Vu", "Vo", "Dang", "Tran", "Dinh", "Trinh", "Ngo", "Dinh"]
    middle_names = ["Van", "Thai", "Duc", "Minh", "Quoc", "Cong", "Huy", "Hong", "Bich", "Thanh", "Dieu", "Thanh", "An", "Hai", "Ngoc"]
    last_names = ["An", "Binh", "Chau", "Dai", "Dung", "Hanh", "Hien", "Khanh", "Lam", "Mai", "Linh", "Trang", "Phuong", "Dat", "Tien"]

    return f"{random.choice(first_names)} {random.choice(middle_names)} {random.choice(last_names)}"

# List of possible genders
genders = ['Nam', 'Nu']

def generate_dob():
    start_date = datetime.date(1970, 1, 1)
    end_date = datetime.date(1990, 12, 31)
    return start_date + (end_date - start_date) * random.random()

# List of possible email domains
email_domains = ['@gmail.com', '@hotmail.com', '@outlook.com']
# Function to generate a random email address
def generate_email(full_name):
    full_name = full_name.replace(' ', '').lower()
    domain = random.choice(email_domains)
    return f'{full_name}{domain}'

def generate_phone_number():
    return '0' + ''.join(random.choices(string.digits, k=10))

def generate_office_id():
    return random.randint(1, 14)

def generate_hire_date():
    start_date = datetime.date(2011, 1, 1)
    end_date = datetime.date(2015, 12, 31)
    return start_date + (end_date - start_date) * random.random()

# Generating tuples
for i in range(1, 2001):
    full_name = generate_vietnamese_name()
    gender = random.choice(genders)
    dob = generate_dob()
    email = generate_email(full_name)
    phone_number = generate_phone_number()
    office_id = generate_office_id()
    hire_date = generate_hire_date()
    
    
    # Print the tuple in the SQL INSERT format, each tuple on a new line
    print(f"('{full_name}', '{gender}', '{dob}', '{email}', '{phone_number}', '{office_id}', '{hire_date}', {i + 8000}),")
```
