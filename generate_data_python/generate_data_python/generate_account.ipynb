```python
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize the Faker library
fake = Faker()

# Function to generate unique login names
def generate_unique_login_names(count, min_length=10, max_length=15):
    login_names = set()
    while len(login_names) < count:
        name = fake.user_name()[:max_length]
        if len(name) >= min_length:
            login_names.add(name)
    return list(login_names)

# Function to generate passwords
def generate_password(min_length=8, max_length=12):
    return fake.password(length=random.randint(min_length, max_length))

# Function to generate random register dates
def generate_register_date(start_year=2016, end_year=2018):
    start_date = datetime(start_year, 1, 1)
    end_date = datetime(end_year, 12, 31)
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return (start_date + timedelta(days=random_days)).date()

# Number of tuples to generate
num_tuples = 10000

# Generate unique login names
login_names = generate_unique_login_names(num_tuples)

# Generate and print tuples
for login_name in login_names:
    password = generate_password()
    register_date = generate_register_date()
    print(f"('{login_name}', '{password}', '{register_date}'),")
```
